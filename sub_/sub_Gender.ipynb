```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style
import seaborn as sns
import unicodedata
```

```python
df = df = pd.read_csv("train.csv")
```

```python
df['Gender']
```

```python
df['Gender'].value_counts()
```

```python
df['Gender'] = df['Gender'].str.replace(' ', '')
df['Gender'] = df['Gender'].str.lower().apply(lambda x: unicodedata.normalize('NFKC', str(x)))
df['Gender'] = df['Gender'].str.normalize('NFKC').str.lower().map({'male': 0, 'female': 1})
```
