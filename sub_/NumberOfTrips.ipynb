```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style
import seaborn as sns

df = pd.read_csv("../data/train.csv")
```

```python
df["NumberOfTrips"].value_counts()
```

```python
def fix_trips(x):
    if pd.isna(x):
        return x
    
    x = str(x)
    if x == '半年に1回':
        return 2.0
    elif x == '四半期に1回':
        return 4.0
    
    # 「年に」「回」などの文字を消して数値化
    x = x.replace('年に', '').replace('回', '')
    return float(x)

# 適用
df["NumberOfTrips"] = df["NumberOfTrips"].apply(fix_trips)
```
