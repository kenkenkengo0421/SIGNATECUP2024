```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style
import seaborn as sns
```

```python
df = pd.read_csv("../data/train.csv")
```

```python
df["Occupation"]
```

```python
df["Occupation"].value_counts()
```

```python
df["Occupation"] = df["Occupation"].str.replace(' ', '').str.lower()
```

```python
df["Occupation"].value_counts()
```

# 全体のデータ

```python
matplotlib.style.use("seaborn-v0_8")
sns.countplot(x='Occupation', data=df)
```

Small Business：中小企業の経営者、自営業、個人事業主、フリーランスなど

Salaried：給与所得者、一般的な会社員（サラリーマン・公務員など）

Large Business：大企業の経営陣、役員、あるいは大企業に勤める高所得なビジネスマン。

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="ProdTaken", hue='Occupation')
plt.xlabel("ProdTaken (0: Not Taken,  1: Taken)")
plt.show()
```

```python
sns.barplot(x='Occupation', y='ProdTaken', data=df, errorbar=None)
```

１） Small Business で成約した人（1）を合計する

２） Small Business の全員の数で割る

３） 成約率になる


棒の高さが「成約率」を表す。純粋に棒が一番高い職業が、一番成約しやすい。
全体の人数（絶対数）こそ少ないが、提案した時に成約してくれる確率（割合）は、Large Businessが一番高い傾向にあると読み取れる。
